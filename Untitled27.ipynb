```python
def fetch_medal_tally(df,year,country):
  medal_df=df.drop_duplicates(subset=["Team","NOC","Games","Year","City","Sport","Event","Medal"])

  flag=0
  if year=="Overall" and country=="Overall":
    temp_df=medal_df
  elif year=="Overall" and country!="Overall":
    flag=1
    temp_df=medal_df[medal_df["region"]==country]
  elif year!="Overall" and country=="Overall":
    temp_df=medal_df[medal_df["Year"]==int(year)]
  else:
    temp_df=medal_df[(medal_df["Year"]==int(year)) & (medal_df["region"]==country)]

  if flag==1:
     x=temp_df.groupby("Year").sum(numeric_only=True)[["Gold","Silver","Bronze"]].reset_index()
  else:
     x=temp_df.groupby("region").sum(numeric_only=True)[["Gold","Silver","Bronze"]].reset_index()
  x["total"]=x["Gold"]+x["Silver"]+x["Bronze"]
  print(x)
```

```python
def most_successfull(df,sport):
  # Use explicit boolean indexing for non-null Medals
  temp_df = df[df["Medal"].notna()]
  print(f"Size of temp_df after filtering for non-null medals: {temp_df.shape}") # Debug print

  if sport!="Overall":
    temp_df=temp_df[temp_df["Sport"]==sport]

  # Get the value counts of names (athletes) who won medals
  # The result of value_counts().reset_index() will have columns 'Name' and 'count'
  athlete_medal_counts = temp_df["Name"].value_counts().reset_index()

  merged_data = athlete_medal_counts.merge(df, left_on="Name", right_on="Name", how="left")

  return merged_data[["Name", "count", "Sport", "region"]].drop_duplicates(subset=["Name"]).rename(columns={"Name": "AthleteName", "count": "Medals"})
```

```python
def most_successfull(df,country):
  temp_df=df.dropna(subset=["Medal"])

  temp_df=temp_df[temp_df["region"]==country]


  athlete_medal_counts = temp_df["Name"].value_counts().reset_index()


  merged_data = athlete_medal_counts.merge(df, left_on="Name", right_on="Name", how="left")


  return merged_data[["Name", "count", "Sport", "region"]].drop_duplicates(subset=["Name"]).rename(columns={"Name": "AthleteName", "count": "Medals"})
```
